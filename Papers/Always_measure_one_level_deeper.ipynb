## Keys to High-Quality Performace Analysis

1. Allow lots of time.

2. Never trust a number generated by a computer.

3. Use your intuition to ask questions, not answer them.

4. **Always measure one level deeper.** If you want to understand the performance of a system at a particular level, you must measure not just that level but also the next level deeper. Measure the underlying factors that contribute to the performance at the higher level. If you are measuring overall latency for remote procedure calls, you could measure deeper by breaking down that latency, determining how much time is spent in the client machine, how much time is spent s in the network, and how much time is spent on the server.
